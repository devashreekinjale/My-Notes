## DataFrames
A tabular data structure with rows and columns(2D).   
Similar to an Excel Spreadsheet

```python
data = {"Name": ["Spongebob", "Patrick", "Squidward"],
        "Age": [30, 35,50]}
```

```python
df = pd.DataFrame(data, index=["Employee 1", "Employee 2", "Employee 3"])
print(df)
```

dataframe with capital D is a constructor

Adding a new column :

```python
df["Job"] = ["Cook", "N/A", "Cashier"]
print(df)
```

Adding a new row :

```python
new_row = pd.DataFrame([{"Name": "Sandy", "Age": 28, "Job": "Engineer"}], index =["Employee 4"])
df = pd.concat([df, new_row])
print(df)
```

Adding new rows :

```python
new_rows = pd.DataFrame([{"Name": "Sandy", "Age": 28, "Job": "Engineer"},
                        {"Name": "Eugene ", "Age": 60, "Job": "Manager"}], index =["Employee 4", "Employee 5"])
df = pd.concat([df, new_rows])
print(df)
```

```python
print(df.loc["Employee 1"]) # or print(df.iloc[0])
```

## Filtering

Keeping the rows that match the condition

```python
df = pd.read_csv("data.csv")
```

```python
tall_pokemon = df[df["Height"]>=2]
print(tall_pokemon)
```

```python
heavy_pokemon = df[df["Weight"]>=100]
print(heavy_pokemon)
```

```python
legendary_pokemon = df[df["Legendary"] == True]
print(legendary_pokemon)
```

```python
water_pokemon = df[(df["Type1"]=="Water") | (df["Type2"]=="Water")]
print(water_pokemon)
```

```python
ff_pokemon = df[(df["Type1"]=="Fire") & (df["Type2"]=="Flying")]
print(ff_pokemon)
```

## Aggregate Functions

Reduces a set of values into a single summary value                 
Used to summarize and analyze data                
Often used with the groupby() function

### 1. Whole dataframe

```python
print(df.mean())
```

we cannot use the mean aggregate function on non-numerics i.e strings

```python
print(df.mean(numeric_only=True)) #finds the mean of any columns that are numeric
```

```python
print(df.sum(numeric_only=True))
```

```python
print(df.min(numeric_only=True))
```

```python
print(df.max(numeric_only=True))
```

```python
print(df.count()) # this does not includes any null values
```

### 2. Single column

```python
print(df["Height"].mean())
```

```python
print(df["Height"].sum())
```

```python
print(df["Height"].min())
```

```python
print(df["Height"].max())
```

```python
print(df["Height"].count())
```

### 3. GroupBy

Grouping by each type and finding the aggregate functions

```python
group = df.groupby("Type1")
```

```python
print(group["Height"].mean())
```

```python
print(group["Height"].sum())
```

```python
print(group["Height"].min())
```

```python
print(group["Height"].max())
```

```python
print(group["Height"].count())
```

## Data Cleaning

The process of fixing / removing: incomplete, incorrect, or irrelevant data     
75% of work done with Pandas is data cleaning

```python
df = pd.read_csv("data.csv")
```

### 1. Drop irrelevant columns

```python
df = df.drop(columns = ["Legendary", "No"])
print(df)
```

### 2. Handle missing data

.dropna() - to drop those rows that are missing a value in the specified column

```python
df = df.dropna(subset=["Type2"])
print(df.to_string())
```

.fillna() - to replace any missing values

```python
df = df.fillna({"Type2": "None"})
print(df.to_string())
```

### 3. Fix inconsistent values

```python
df["Type1"] = df["Type1"].replace({"Grass": "GRASS"})
print(df.to_string())
```

```python
df["Type1"] = df["Type1"].replace({"Grass": "GRASS",
                                    "Fire": "FIRE",
                                    "Water": "WATER"})
print(df.to_string())
```

### 4. Standardize Text

```python
df["Name"] = df["Name"].str.lower()
print(df.to_string())
```

```python
df["Name"] = df["Name"].str.upper()
print(df.to_string())
```

### 5. Fix data types

```python
df = pd.read_csv("data.csv")
```

```python
df["Legendary"] = df["Legendary"].astype(bool)
print(df)
```

### 6. Remove duplicate values

```python
df = pd.read_csv("data.csv")
print(df)
```

```python
df = df.drop_duplicates()
print(df.to_string())
```
