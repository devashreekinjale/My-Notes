# Matplotlib

## Intro

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =np.array([2023,2024,2025,2026])
y = np.array([15, 25, 30, 20])
plt.plot(x,y)
plt.show()
```

## Plot Customization

1. Dot (.)

```python
plt.plot(x,y, marker = ".", 
        markersize = 20, 
        markerfacecolor = "#1cd3fc", 
        markeredgecolor = "#1cd3fc",
        linestyle = "solid",
        linewidth = 2,
        color = "#1c5bfc")
plt.show()
```

2. Circle (o)

```python
plt.plot(x,y, marker = "o")
plt.show()
```

3. Inverted Triangle (v)

```python
plt.plot(x,y, marker = "v")
plt.show()
```

4. Star (*)

```python
plt.plot(x,y, marker = "*")
plt.show()
```

---

```python
x =np.array([2023,2024,2025,2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17,23,38,5])
```

```python
line_style = dict(marker = ".", 
        markersize = 20, 
        markerfacecolor = "#1cd3fc", 
        markeredgecolor = "#1cd3fc",
        linestyle = "solid",
        linewidth = 2,
        color = "#1c5bfc")
```

**line_style - unpacking the linestyle dictionary

```python
plt.plot(x,y1, **line_style)
plt.plot(x,y2, **line_style)
plt.show()
```

---

For different colors :

```python
line_style = dict(marker = ".", 
        markersize = 20, 
        markerfacecolor = "#1cd3fc", 
        markeredgecolor = "#1cd3fc",
        linestyle = "solid",
        linewidth = 2)

plt.plot(x,y1,color ="blue", **line_style)
plt.plot(x,y2,color ="red", **line_style)
plt.show()
```

## Labels

```python
x =np.array([2023,2024,2025,2026])
y1 = np.array([15,25,30,20])
y2 = np.array([17,23,38,5])
y3 = np.array([13,15,20,30])

plt.title("Class Size", fontsize = 25, 
            family = "Arial",
            fontweight = "bold",
            color = "purple")

plt.xlabel("Year", fontsize = 15,
            family = "Arial",
            fontweight = "bold",
            color = "#a305b5")

plt.ylabel("Students", fontsize = 15,
            family = "Arial",
            fontweight = "bold",
            color = "#a305b5")

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

.xticks() - to have the scale on x axis same as the variables in our array

```python
x =np.array([2023,2024,2025,2026])
y1 = np.array([15,25,30,20])
y2 = np.array([17,23,38,5])
y3 = np.array([13,15,20,30])

plt.title("Class Size", fontsize = 25, 
            family = "Arial",
            fontweight = "bold",
            color = "purple")

plt.xlabel("Year", fontsize = 15,
            family = "Arial",
            fontweight = "bold",
            color = "#a305b5")

plt.ylabel("Students", fontsize = 15,
            family = "Arial",
            fontweight = "bold",
            color = "#a305b5")

plt.tick_params(axis = "both",
                colors = "#a305b5" )

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xticks(x)
plt.show()
```

## Bar Graphs
Compares categories of data by representing each category with a bar

```python
categories =np.array(["Grains", "Fruits", "Vegetables", "Protiens", "Dairy", "Sweets"])
values = np.array([4,3,2,5,3,1])
```

```python
plt.bar(categories, values, color = "skyblue")
plt.show()
```

```python
plt.barh(categories, values, color = "skyblue") # this creates a horizontal bar graph
plt.show()
```

```python
plt.bar(categories, values, color = "skyblue")
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```

## Pie Charts
Circular chart divided into slices to show percentages of the total.        
Good for visualizing distribution among categories.

```python
categories = ["Freshman", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])
```

```python
plt.pie(values, labels = categories,
        autopct="%1.1f%%",
        explode=[0,0,0,0.2],
        shadow = True,
        startangle = 90)

plt.title("College", fontsize = 20)
plt.show()
```

1.1 to display the decimal values having only one digit after the decimal             
% : single one for formating and double for displaying the sign           
**explode** is used to seperate the parts of pie chart           
shadow = True to give the pie a shadow          
startangle is to rotate the pie chart

```python
colors =["red","yellow", "blue", "green"]
plt.pie(values, labels = categories,
        autopct="%1.1f%%",
        colors = colors,
        explode=[0,0,0,0.2],
        shadow = True,
        startangle = 90)

plt.title("College", fontsize = 20)
plt.show()
```

## Scatter Graphs
Shows the relationship between two variables           
Helps to identify a correlation (+,-,None)           
Example: Study hours vs. Test scores

```python
x = np.array([0,1,1,2,3,4,5,6,7,7,8]) # Hours Studied
y = np.array([55,60,65,62,68,70,75,78,82,85,87]) # Grades
```

alpha is the transpareny of our markers i.e dots        
s is for the size of the dots

```python
plt.scatter(x,y, color = "skyblue",
            alpha = 0.5,
            s = 100)

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grades")

plt.show()
```

.legend() - a box appears which tells us which data is of which

```python
x1 = np.array([0,1,1,2,3,4,5,6,7,7,8]) # Hours Studied
y1 = np.array([55,60,65,62,68,70,75,78,82,85,87]) # Grades

x2 = np.array([0,1,2,2,3,4,5,6,7,8,8]) # Hours Studied
y2 = np.array([50,58,65,70,72,78,83,88,92,95,97]) # Grades

plt.scatter(x1,y1, color = "skyblue",
            alpha = 0.5,
            s = 100,
            label = "Class A")

plt.scatter(x2,y2, color = "red",
            alpha = 0.5,
            s = 100,
            label = "Class B")

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grades")

plt.legend() 
plt.show()
```

## Histograms
A visual representation of the distribution of quantitative data.           
They group values into bins (intervals)
and counts how many fall into each range.

```python
scores = np.random.normal(loc=80, scale = 10, size = 100)
```

```python
plt.hist(scores)
plt.show()
```

----

to ensure no score goes bellow 0 or above 100, we will use clip()

```python
scores = np.random.normal(loc=80, scale = 10, size = 100)
```

```python
scores = np.clip(scores, 0, 100)
plt.hist(scores)
plt.show()
```

```python

plt.hist(scores, bins = 10,
                color = "skyblue",
                edgecolor = "black")

plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("No. of students")
plt.show()
```

## Subplots

1. Figure = The entire canvas 
2. Ax = A single plot (subplot) 

```python
print(plt.subplots(2,2))
```

```python
figure, axes = plt.subplots(2,2)
```

```python
x = np.array([1, 2, 3, 4, 5])
figure, axes = plt.subplots(2,2)

axes[0,0].plot(x, x*2, color = "red")
axes[0,0].set_title("x*2")

plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
figure, axes = plt.subplots(2,2)

axes[0,0].plot(x, x*2, color = "red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x, x**2, color = "blue")
axes[0,1].set_title("x**2")

plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
figure, axes = plt.subplots(2,2)

axes[0,0].plot(x, x*2, color = "red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x, x**2, color = "blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(x, x**3, color = "green")
axes[1,0].set_title("x**3")

plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
figure, axes = plt.subplots(2,2)

axes[0,0].plot(x, x*2, color = "red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x, x**2, color = "blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(x, x**3, color = "green")
axes[1,0].set_title("x**3")

axes[1,1].plot(x, x**4, color = "brown")
axes[1,1].set_title("x**4")

plt.tight_layout() # to fit in place
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
figure, axes = plt.subplots(2,2)

axes[0,0].plot(x, x*2, color = "red")
axes[0,0].set_title("x*2")

axes[0,1].bar(x, x**2, color = "blue")
axes[0,1].set_title("x**2")

axes[1,0].barh(x, x**3, color = "green")
axes[1,0].set_title("x**3")

axes[1,1].plot(x, x**4, color = "brown")
axes[1,1].grid()
axes[1,1].set_title("x**4")

plt.tight_layout() # to fit in place
plt.show()
```

## Pandas + Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
print(df)
```

```python
print(df["Type1"].value_counts())
```

```python
type_count = df["Type1"].value_counts()

plt.bar(type_count.index, type_count.values)
plt.show()
```

```python
type_count = df["Type1"].value_counts()

plt.barh(type_count.index, type_count.values)
plt.show()
```

```python
type_count = df["Type1"].value_counts(ascending = True)

plt.barh(type_count.index, type_count.values, color= "pink", edgecolor = "purple")

plt.title("No. of Pokemon by Primary Type")
plt.xlabel("Count")
plt.ylabel("Type")
plt.tight_layout()
plt.show()
```
